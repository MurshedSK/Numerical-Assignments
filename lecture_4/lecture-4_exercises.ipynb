We are now going to focus on the integration interval $[0,3]$ and compare the absolute error of the 5 rules for an increasing number of sub-intervals $n$ (go up to $n=10000$).

Plot your results (error versus number of sub-intervals) and comment.

````python
def left_rule(a, b, n):
    h = (b - a) / n
    x = np.linspace(a, b, n, endpoint=False)
    return h * np.sum(f(x))

def right_rule(a, b, n):
    h = (b - a) / n
    x = np.linspace(a + h, b, n)
    return h * np.sum(f(x))

def midpoint_rule(a, b, n):
    h = (b - a) / n
    x = np.linspace(a + h/2, b - h/2, n)
    return h * np.sum(f(x))

def trapezoidal_rule(a, b, n):
    h = (b - a) / n
    x = np.linspace(a, b, n+1)
    return h * (0.5 * f(a) + 0.5 * f(b) + np.sum(f(x[1:-1])))

def simpson_rule(a, b, n):
    if n % 2 != 0:
        raise ValueError("n must be even for Simpson's rule")
    h = (b - a) / n
    x = np.linspace(a, b, n+1)
    return h/3 * (f(a) + f(b) + 4*np.sum(f(x[1:-1:2])) + 2*np.sum(f(x[2:-1:2])))

def abs_error(numerical_value, actual_value):
    return abs(numerical_value - actual_value)


from scipy import integrate

# Calculate the "exact" value using SciPy's quad function
exact_value, _ = integrate.quad(f, 0, 1)

print(f"Testing numerical integration methods for x^2 * exp(-x^2) dx from 0 to 1")
print(f"'Exact' value (using SciPy's quad): {exact_value:.10f}")

for n in [10, 100, 1000]:
    print(f"\nNumber of subintervals: {n}")

    left = left_rule(0, 1, n)
    right = right_rule(0, 1, n)
    mid = midpoint_rule(0, 1, n)
    trap = trapezoidal_rule(0, 1, n)
    simp = simpson_rule(0, 1, n)

    print(f"Left Rectangle:   {left:.10f}, Error: {abs_error(left, exact_value):.10f}")
    print(f"Right Rectangle:  {right:.10f}, Error: {abs_error(right, exact_value):.10f}")
    print(f"Midpoint:         {mid:.10f}, Error: {abs_error(mid, exact_value):.10f}")
    print(f"Trapezoidal:      {trap:.10f}, Error: {abs_error(trap, exact_value):.10f}")
    print(f"Simpson's:        {simp:.10f}, Error: {abs_error(simp, exact_value):.10f}") 
````

## Monte Carlo Integration

In this exercise you are going to:

- implement a Monte Carlo integration routine 
- find the integral of $f(x)=x^2$ over the interval $[-2,2]$ 
- study the dependence of the absolute error on the number of samples.

Define the function $f(x)$ and a function `montecarlo` that returns the integral.

```python
def montecarlo(f,a,b,n):
    """f is function to integrate over the interval [a,b] with n random points"""
    ...
    ...
    return I
```

````python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

def montecarlo(func, a, b, n):
    """
    Monte Carlo integration of func over [a, b] with n samples.
    """
    x = np.random.uniform(a, b, n)
    y = func(x)
    integral = (b - a) * np.mean(y)
    return integral

# Exact value of the integral of x^2 from -2 to 2
exact_value = 16/3  # (2^3 - (-2)^3) / 3

# Range of sample sizes to test
n_samples = np.logspace(2, 6, 50).astype(int)

# Perform Monte Carlo integration for each sample size
results = [montecarlo(f, -2, 2, n) for n in n_samples]

# Calculate absolute errors
errors = np.abs(np.array(results) - exact_value)

# Plot the results
plt.figure(figsize=(10, 6))
plt.loglog(n_samples, errors, 'b.', label='Monte Carlo')
plt.loglog(n_samples, 1/np.sqrt(n_samples), 'r--', label='1/n')
plt.xlabel('Number of samples')
plt.ylabel('Absolute Error')
plt.title('Monte Carlo Integration: Error vs. Number of Samples')
plt.legend()
plt.grid(True)
plt.show()

# Print some example results
print(f"Exact value: {exact_value}")
for n in [100, 1000, 10000, 100000]:
    result = montecarlo(f, -2, 2, n)
    error = abs(result - exact_value)
    print(f"n = {n}: Integral  {result:.6f}, Absolute Error = {error:.6f}")
````

Study the dependence of the absolute error on the number of random samples $n$ with $n=10\dots10^4$

- For each value of $n$ calculate the integral $N=100$ times and take the mean value $\left< I(n)\right>$.
- Calculate the absolute error with respect to this mean, plot it as a function of $n$.
- Fit the data and remember that the error is expected to be
    $$E \sim \frac{1}{\sqrt{n}}$$
    
- For $N=1000$ and $n=1000$, plot the probability distribution function of $\left< I(n)\right>$. Comment. 

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def f(x):
    return x**2

def monte_carlo_integration(func, a, b, n):
    x = np.random.uniform(a, b, n)
    y = func(x)
    integral = (b - a) * np.mean(y)
    return integral

# Exact value of the integral
exact_value = 16/3

# Range of sample sizes
n_samples = np.logspace(1, 4, 20).astype(int)

# Number of repetitions for each sample size
N = 100

# Calculate mean integral values and errors
mean_integrals = []
abs_errors = []

for n in n_samples:
    integrals = [monte_carlo_integration(f, -2, 2, n) for _ in range(N)]
    mean_integral = np.mean(integrals)
    mean_integrals.append(mean_integral)
    abs_errors.append(np.abs(mean_integral - exact_value))

# Fit the error data
def error_model(n, a):
    return a / np.sqrt(n)

popt, _ = curve_fit(error_model, n_samples, abs_errors)

# Plot the results
plt.figure(figsize=(10, 6))
plt.loglog(n_samples, abs_errors, 'b.', label='Monte Carlo')
plt.loglog(n_samples, error_model(n_samples, *popt), 'r-', label='Fit: a/n')
plt.xlabel('Number of samples (n)')
plt.ylabel('Absolute Error')
plt.title('Monte Carlo Integration: Error vs. Number of Samples')
plt.legend()
plt.grid(True)
plt.show()

# For N=1000 and n=1000, plot the probability distribution
N_large = 1000
n_large = 1000
integrals_large = [monte_carlo_integration(f, -2, 2, n_large) for _ in range(N_large)]

plt.figure(figsize=(10, 6))
plt.hist(integrals_large, bins=30, density=True, alpha=0.7)
plt.axvline(exact_value, color='r', linestyle='dashed', linewidth=2, label='Exact Value')
plt.xlabel('Integral Value')
plt.ylabel('Probability Density')
plt.title(f'Distribution of Monte Carlo Integral Estimates (N={N_large}, n={n_large})')
plt.legend()
plt.grid(True)
plt.show()

print(f"Fitted parameter a: {popt[0]:.4f}")
print(f"Mean of large sample: {np.mean(integrals_large):.6f}")
print(f"Standard deviation of large sample: {np.std(integrals_large):.6f}")
````
